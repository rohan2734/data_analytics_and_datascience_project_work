### standard deviation
- meaning is, dispersion of dataset from mean
- what is the significance with standard deviation? when SD is high or SD is low.
- if SD is very high, then what is the effect on the top of the distribution of data, it will affect the ML model to learn the entire pattern, if SD is very less, also it will affect the model

### variances

### 10.from a cabin colum ,seperate text and numeric value

```python
def seperateTextAndNumber(string):
    text=""
    numeric=""
    for i in string:
#         if type(i)
#         print(type(i))
        if i >='0' and i <='9':
            numeric+=i
        elif i == ' ':
            pass
        else:
            text+=i
            
    return text,numeric
```

```python
seperateTextAndNumber("C23 C25 C27")
```
