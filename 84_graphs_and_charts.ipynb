## Inter quartile Range 
- range means value of the dataset whic is available in between
- q2 is median(50th percentile) all the time
- q3 is 75%(75th percentile), q1(25th percentile)
- it calculates minimum   by q1-1.5 * (iqr)
- we can find the maximum by q3 + 1.5*(iqr)

## 3 column comparision

```python
df.plot.scatter(x='SepalLengthCm',y='PetalWidthCm',c='SepalWidthCm',s=20)
#c is color, and s is size 
```

### obervation
- x axis is SepalLengthCm, y axis is PetalWidthCm , c is color axis is SepalWidthCm
- color of each dot is not same.
- color coding will start from 2 to 4.5
#### how is it deciding the color of the dot ??
- it finds sepal length and petal length, and based on sepalWidth , it will give the color


```python
df.plot.scatter(x='SepalLengthCm',y='PetalWidthCm',c='SepalWidthCm',s=df['PetalWidthCm']*100)
```
