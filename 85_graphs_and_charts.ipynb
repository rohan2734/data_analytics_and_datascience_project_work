## Tips Dataset

```python
df1=sns.load_dataset('tips')
df1
```

- total bill,tip, male or female, which day you are going, are you going for dinner or not, and number of people going etc

## relationship between total_bill and tips

```python
## relationship between total bill and tips
sns.scatterplot(df1.total_bill,df1.tip)
```

## rel plot - relationplot

```python
sns.relplot(x='total_bill',y='tip',data=df1)
```

### observation
- if total bills are increasing  , tips are increasing,
- if total bills are very high, people are able to get very high tips
- there is some sort of relationship , if total bills are less, tips are less 
- there is also a chance where, bills are less, but  still tips are high 

```python
### how many people are smoker or how many people are non smoker
df1['smoker'].value_counts()
# if restaurants are receiving so many smokers, then they can make a arrangement for them
```

## relplot , hue
- hue does group by operation and represents different groups in different colors

```python
sns.relplot(x='total_bill',y='tip',data=df1,hue='smoker')
```

```python
- highest tips are smoker 
```

### style - relplot
- based on different groups, it will give different shapes

```python
sns.relplot(x='total_bill',y='tip',data=df1,style='smoker')
```

```python
sns.relplot(x='total_bill',y='tip',data=df1,style='size')
```

### relplot - style and hue

```python
sns.relplot(x='total_bill',y='tip',data=df1,style='time',hue='smoker')
## group by operations based on time and smoker columns
```

### observation
- person who paid lowest bill, went for dinner, and also a smoker

### relplot->Size

```python
sns.relplot(x='total_bill',y='tip',data=df1,size='size')
```

### relplot-line

```python
sns.relplot(x='total_bill',y='tip',kind='line',data=df1)
```

## Categorical Plot 

```python
sns.catplot(x='day',y='total_bill',data=df1)
#we can see that on saturday, people ahve paid the highest bill.
# on friday, there are very few people , it will help the restaurant to make the decision
```

### catplot - swarm
- it doesnt give much additional insights 

```python
sns.catplot(x='day',y='tip',data=df1,kind='swarm')
```

### catplot - Box

```python
sns.catplot(x='day',y='tip',kind='box',data=df1)
```

### observation
- exceptionally well tips , persons have received on saturday and thursday - outliers 
- outliers - exceptional things, if we have values high than 1.5* the q3.

```python
sns.catplot(x='day',y='tip',kind='box',data=df1,hue='sex')
```
